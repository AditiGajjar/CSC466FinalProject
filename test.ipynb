```python
def normalize_column(column):
    return (column - column.min()) / (column.max() - column.min())

path = "Spotify_data.xlsx"
dfin = pd.read_excel(path)
df = dfin[['Age', 'Gender', 'spotify_usage_period', 'preferred_listening_content',
       'spotify_subscription_plan', 'music_time_slot', 'music_Influencial_mood',
       'music_lis_frequency', 'music_expl_method', 'music_recc_rating', 'fav_music_genre']]
genres = df["fav_music_genre"].unique()
numgenres = len(genres)


dumcols = ['music_Influencial_mood','music_lis_frequency', 'music_expl_method']
for col in dumcols:
    df_dummies = df[col].str.get_dummies(',')
    df_dummies.columns = df_dummies.columns.str.strip()
    df_dummies = df_dummies.add_prefix(col+"_")

    # Concatenate the dummy columns with the original DataFrame
    df = pd.concat([df, df_dummies], axis=1)
    df.drop(col, axis=1, inplace=True)

df["music_recc_rating"] = normalize_column(df["music_recc_rating"])



dumm = pd.get_dummies(df) * 1
dumm.columns = dumm.columns.str.replace(" ", "_")

dumm.to_csv("dummified_user.csv", index=False)
dumm.columnsimport pandas as pd
import numpy as np

# Defining the genres
genres = ["EDM", "Country", "Pop", "HipHop", "Film", "Classical", "Religious", "Latin", "Rock", "Jazz", "R&B", "Soul", "Folk"]

# Creating a DataFrame to hold the dataset

# Populating the dataset
np.random.seed(0)  # For reproducibility
n = 152
    # Base ratings with some randomness
ratings = {
    "EDM": np.random.normal(4, 2, n).astype(int),  # Mean: 7, Standard Deviation: 2
    "Country": np.random.normal(5, 2, n).astype(int),  # Mean: 5, Standard Deviation: 2
    "Pop": np.random.normal(7.5, 2, n).astype(int),  # Mean: 6, Standard Deviation: 2
    "HipHop": np.random.normal(8, 4, n).astype(int),  # Mean: 8, Standard Deviation: 1.5
    "Film": np.random.normal(5, 2, n).astype(int),  # Mean: 5, Standard Deviation: 2
    "Classical": np.random.normal(5, 2, n).astype(int),  # Mean: 3, Standard Deviation: 1.5
    "Religious": np.random.normal(4, 2, n).astype(int),  # Mean: 4, Standard Deviation: 2
    "Latin": np.random.normal(6, 1.5, n).astype(int),  # Mean: 6, Standard Deviation: 1.5
    "Rock": np.random.normal(7, 1.5, n).astype(int),  # Mean: 7, Standard Deviation: 1.5
    "Jazz": np.random.normal(5, 2, n).astype(int),  # Mean: 5, Standard Deviation: 2
    "R&B": np.random.normal(6, 2, n).astype(int),  # Mean: 6, Standard Deviation: 2
    "Soul": np.random.normal(5, 2, n).astype(int),  # Mean: 5, Standard Deviation: 2
    "Folk": np.random.normal(4, 1.5, n).astype(int) 
}

# Adjusting ratings based on genre affinity
m1 = ratings["Pop"] > 7
ratings["HipHop"][m1] += 1
ratings["R&B"][m1] += 1
ratings["Folk"] -= 1
ratings["Classical"] -= 1

m2 = ratings["Classical"] > 5
ratings["Film"][m2] += 1
ratings["Jazz"][m2] += 2

m3 = ratings["Rock"] > 6
ratings["EDM"][m3] += 1
ratings["Folk"][m3] += 1

m4 = ratings["Jazz"] > 7
ratings["Pop"][m4] += 1
ratings["R&B"][m4] += 1
ratings["Folk"][m4] -= 1
ratings["Rock"][m4] -= 1

m5 = ratings["Rock"] > 8
ratings["EDM"][m5] += 2
ratings["Pop"][m5] -= 1
ratings["Classical"][m5] -= 1

m6 = ratings["HipHop"] > 6
ratings["Rock"][m6] += 1
ratings["Jazz"][m6] += 1
ratings["Folk"][m6] -= 1
ratings["Film"][m6] -= 1

# Ensuring ratings are within 1-10
sd = np.zeros(numgenres)
for i, genre in enumerate(genres):
    ratings[genre] = np.minimum(10, ratings[genre])
    ratings[genre] = np.maximum(1, ratings[genre])
    sd[i] = np.std(ratings[genre])


# Adding the row to the dataset
dataset = pd.DataFrame(ratings)

dataset.to_csv("spotify_ratings.csv", index=False)
```
